```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
cols=["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal","num"]
df=pd.read_csv('Cleavland.csv')
df.columns=cols
df
```

# a. Data cleaning(Remove NA, ?, Negative values etc.)

```python
df.isnull().sum()
```

```python
df1=df.dropna()
df1.isnull().sum()
```

```python
df2=df1.replace('?',pd.NA)
df2
```

```python
df2.isnull().sum()
```

```python
df3=df2.dropna()
df3
```

# Error correcting(Outlier detection and removal)

```python
df3.std()
```

```python
df3.mean()
```

```python
Out=np.abs((df3-df3.mean())/df3.std())
lier=Out>3
print(lier.sum())
df3=df3[~lier.any(axis=1)]
df3
```

# Data transformation

```python
df3.loc[df3['sex']==1,'sex']='M'
df3.loc[df3['sex']==0,'sex']='F'
lb=LabelEncoder()
df3['sex']=lb.fit_transform(df3['sex'])
df3
```

# Build Data model using regression and kNN methods and compare accuracy of heart disease prediction.

```python
X=df3.drop('num',axis=1)
Y=df3['num']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
r1=reg.predict(x_test)
acc1=accuracy_score(y_test,r1)
print(acc1)
```

```python
reg=KNeighborsClassifier()
reg.fit(x_train,y_train)
```

```python
r2=reg.predict(x_test)
acc2=accuracy_score(y_test,r2)
print(acc2)
```
