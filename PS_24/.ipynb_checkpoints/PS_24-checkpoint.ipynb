```python
import pandas as pd
```

```python
df=pd.read_csv('forestfires.csv')
df
```

```python
notaffected=df[df['area']==0.00]
notaffected
```

```python
partially=df[(df['area']!=0.00) & (df['area']<=0.50)]
partially
```

```python
fully=df[df['area']>0.50]
fully
```

```python
merge1=pd.concat([notaffected,fully])
merge1
```
